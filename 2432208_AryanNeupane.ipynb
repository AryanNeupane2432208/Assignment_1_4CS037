```python
# • Adding New Columns:Happpiness_category
def categorize_happiness(score):
    if score < 4:
        return 'Low'
    elif 4 <= score <= 6:
        return 'Medium'
    else:
        return 'High'

# Create a new column 'Happiness Category' using .apply()
whr_df['Happiness Category'] = whr_df['score'].apply(categorize_happiness)

# Print the DataFrame with the new column
whr_df
```

3.2 Problem - 2 - Some Advance Data Exploration Task:

```python
# Task - 1 - Setup Task - Preparing the South-Asia Dataset:

south_asian_countries = ["Afghanistan", "Bangladesh", "Bhutan", "India","Maldives", "Nepal", "Pakistan","Sri Lanka"]

south_asian_data = whr_df[whr_df['Country name'].isin(south_asian_countries)]
south_asian_data.head(10)
```

```python
south_asian_data.to_csv('South_Asian_WHR.csv',index=False)
south_asian_df = pd.read_csv('South_Asian_WHR.csv')
south_asian_df
```

```python
south_asian_df.describe()
```

```python
# Task - 2 - Composite Score Ranking:

# making new column composite score
south_asian_df['Composite Score']=(0.40 * south_asian_df['Log GDP per capita'] +
                                  0.30 * south_asian_df['Social support'] +
                                  0.30 * south_asian_df['Healthy life expectancy'])

# sorting the data
sorted_south_asian_df = south_asian_df.sort_values(by='Composite Score', ascending=False)
sorted_south_asian_df.head()
```

```python
# 3. Visualize the top 5 countries using a horizontal bar chart showing the Composite Score.

# taking top 5 countries
top_5_sorted_south_asian_df = sorted_south_asian_df.head(5)

#a bar plot for visulization acc to composite score
plt.figure(figsize=(10, 6))
sns.barplot(y='Country name', x='Composite Score', data=top_5_sorted_south_asian_df,color='steelblue',edgecolor='black')
plt.ylabel('Country')
plt.xlabel('Composite Score')
plt.title('Top 5 Countries by Composite Score')
plt.show()
```

```python
# comparing the original ranking acc to score and composite ranking acc to composite score

# Plot the Original Score as a line plot
plt.figure(figsize=(8, 6))
plt.plot(south_asian_df['Country name'],south_asian_df['score'],color='red',marker='o',linestyle='-',label='Original Score')

# Plot the Composite Score as a line plot
plt.plot(south_asian_df['Country name'],south_asian_df['Composite Score'],color='green',marker='o',linestyle='-',label='Composite Score')

# Set plot labels and title
plt.xlabel('Countries', fontsize=14)
plt.ylabel('Scores', fontsize=14)
plt.title('Original Score vs Composite Score', fontsize=16)

# Rotate x-axis labels for better readability and adjust spacing
plt.xticks(rotation=0, fontsize=12)
plt.yticks(fontsize=12)

#grid lines for better score comparison
plt.grid(axis='y', linestyle='--')
plt.legend(fontsize=12, loc='upper right')
# Add padding around the plot
plt.tight_layout()
# Show the plot
plt.show()
```

```python
# Task - 3 - Outlier Detection:

# for outliers acc to gdp
Q1_GDP = south_asian_df['Log GDP per capita'].quantile(0.25)
Q3_GDP = south_asian_df['Log GDP per capita'].quantile(0.75)
IQR_GDP = Q3_GDP - Q1_GDP

above_outlier_gdp =Q3_GDP + 1.5 * IQR_GDP
below_outlier_gdp=Q1_GDP - 1.5 * IQR_GDP

outlier_country_gdp = south_asian_df[(south_asian_df['Log GDP per capita'] > above_outlier_gdp) | (south_asian_df['Log GDP per capita'] < below_outlier_gdp)]

# for outliers acc to score
Q1_score = south_asian_df['score'].quantile(0.25)
Q3_score = south_asian_df['score'].quantile(0.75)
IQR_score = Q3_score - Q1_score

above_outlier_score =Q3_score + 1.5 * IQR_score
below_outlier_score=Q1_score - 1.5 * IQR_score

outlier_country_score = south_asian_df[(south_asian_df['score'] > above_outlier_score) | (south_asian_df['score'] < below_outlier_score)]

# Create a new column 'Outlier' initialized as 'Normal'
south_asian_df['Outlier'] = 'Normal'

# Mark outliers in GDP or Score as 'Outlier'
south_asian_df.loc[
    (south_asian_df['Log GDP per capita'] > above_outlier_gdp) |
    (south_asian_df['Log GDP per capita'] < below_outlier_gdp) |
    (south_asian_df['score'] > above_outlier_score) |
    (south_asian_df['score'] < below_outlier_score),
    'Outlier'
] = 'Outlier'

south_asian_df[['Country name','score','Log GDP per capita','Outlier']]
```

```python
# Scatter plot with outliers highlighted
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Log GDP per capita',y='score',hue='Outlier',data=south_asian_df,s=100)

# Annotate points with country names
for i, row in south_asian_df.iterrows():
        plt.text(
            row['Log GDP per capita']-0.015,  # x
            row['score'] -0.16,               # y
            row['Country name'],              # text to write
            fontsize=9,
            color='black'
        )

# Add labels and title
plt.title('Scatter Plot of Log GDP per Capita vs Score (Highlighting Outliers)', fontsize=14)
plt.xlabel('Log GDP per Capita', fontsize=12)
plt.ylabel('Score', fontsize=9)
plt.legend(title='Data Points', fontsize=10)
plt.grid(alpha=0.3)
plt.show()
```

```python
# Task - 4 - Exploring Trends Across Metrics:

# 1. Calculate Pearson correlation
social_corr = south_asian_df['Social support'].corr(south_asian_df['score'])
generosity_corr = south_asian_df['Generosity'].corr(south_asian_df['score'])

print(f"Correlation of Social support with score: {social_corr}")
print(f"Correlation of Generosity with score    : {generosity_corr}")

# 2. Create scatter plots with trendlines
metrics = ['Social support', 'Generosity']
for metric in metrics:
    plt.figure(figsize=(8, 5))
    sns.regplot(x=metric, y='score', data=south_asian_df,scatter=True)
    plt.title(f"Scatter Plot with Trendline: {metric} vs Score")
    plt.xlabel(metric)
    plt.ylabel('Score')
    plt.grid(axis='y',linestyle='--')
    plt.show()

# 3. Analyze strongest and weakest relationships
if abs(social_corr) > abs(generosity_corr):
    strongest_metric = 'Social support'
    weakest_metric = 'Generosity'
else:
    strongest_metric = 'Generosity'
    weakest_metric = 'Social support'

print(f"Strongest Relationship: {strongest_metric} with Score")
print(f"Weakest Relationship: {weakest_metric} with Score")
```

```python
# Task - 5 - Gap Analysis:


# 1. Add a new column, GDP-Score Gap
south_asian_df['GDP-Score Gap'] = abs(south_asian_df['Log GDP per capita'] - south_asian_df['score'])

# 2. Rank the countries by the gap (ascending and descending)
ranked_ascending = south_asian_df.sort_values(by='GDP-Score Gap', ascending=True)
ranked_descending = south_asian_df.sort_values(by='GDP-Score Gap', ascending=False)

# Print top results
print("Ranked by GDP-Score Gap (Ascending):")
print(ranked_ascending[['Country name', 'GDP-Score Gap']].head())

print("\nRanked by GDP-Score Gap (Descending):")
print(ranked_descending[['Country name', 'GDP-Score Gap']].head())

# 3. Highlight the top 3 countries with the largest positive and negative gaps using a bar chart

# Extract top positive and negative gaps
top_positive = ranked_descending.head(3)
top_negative = ranked_ascending.head(3)

# Bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='GDP-Score Gap', y='Country name', data=top_positive, color='lightgreen', label='Largest Gaps')
sns.barplot(x='GDP-Score Gap', y='Country name', data=top_negative, color='steelblue', label='Smallest Gaps')
plt.title('Top 3 Countries with Largest and Smallest GDP-Score Gaps')
plt.xlabel('GDP-Score Gap')
plt.ylabel('Country')
plt.legend(title='Gap Type')
plt.tight_layout()
plt.show()
```

```python
# Task - 1 - Setup Task - Preparing the Middle Eastern Dataset:

middle_east_countries = [ "Bahrain", "Iran", "Iraq", "Israel", "Jordan",
"Kuwait", "Lebanon", "Oman", "State of Palestine", "Qatar", "Saudi Arabia", "Syria","United Arab Emirates", "Yemen"]
middle_east_data = whr_df[whr_df['Country name'].isin(middle_east_countries)]

middle_east_data.to_csv('middle_east_data.csv',index=False)
middle_east_df = pd.DataFrame(pd.read_csv('middle_east_data.csv'))

middle_east_df
```

```python
# 1. Descriptive Statistics:
south_asia_mean = south_asian_df['score'].mean()
south_asia_std = south_asian_df['score'].std()

middle_east_mean = middle_east_df['score'].mean()
middle_east_std = middle_east_df['score'].std()

print(f"South Asia Mean Score: {south_asia_mean}")
print(f"South Asia Standard Deviation: {south_asia_std}")

print(f"Middle East Mean Score: {middle_east_mean}")
print(f"Middle East Standard Deviation: {middle_east_std}")
```

```python
# 2. Top and Bottom Performers:
south_asia_top_3 = south_asian_df.nlargest(3, 'score')
south_asia_bottom_3 = south_asian_df.nsmallest(3, 'score')

middle_east_top_3 = middle_east_df.nlargest(3, 'score')
middle_east_bottom_3 = middle_east_df.nsmallest(3, 'score')

# top performs bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='score', y='Country name', data=south_asia_top_3, color='lightgreen', label='South Asia Top 3')
sns.barplot(x='score', y='Country name', data=middle_east_top_3, color='steelblue', label='Middle east Top 3')
plt.xlabel('Score')
plt.ylabel('Country')
plt.title('Top 3 Performers in South Asia and Middle East')
plt.legend()
plt.show()

# bottom performers bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='score', y='Country name', data=south_asia_bottom_3, color='lightgreen', label='South Asia Bottom 3')
sns.barplot(x='score', y='Country name', data=middle_east_bottom_3, color='steelblue', label='Middle east Bottom 3')
plt.xlabel('Score')
plt.ylabel('Country')
plt.title('Bottom 3 Performers in South Asia and Middle East')
plt.legend()
plt.show()

# south_asian_df
```

```python
# 3. Metric Comparisons:

# adding respective regions in tha dataframes
south_asian_df['Region'] = 'South Asia'
middle_east_df['Region'] = 'Middle East'

# Combining data from both regions
combined_plot_df = (pd.concat([south_asian_df, middle_east_df])).groupby('Region')[['Log GDP per capita', 'Social support', 'Healthy life expectancy']].mean()

combined_plot_df.head(10)

combined_plot_df.plot(kind='bar')
plt.xlabel('Region')
plt.ylabel('Average Score')
plt.title('Comparision on key metrics of two reigions')
plt.xticks(rotation=0)
plt.show()
```

```python
# 4. Happiness Disparity:
range_south_asia = south_asian_df['score'].max() - south_asian_df['score'].min()
cv_south_asia = (range_south_asia/south_asia_mean)*100

range_middle_east = middle_east_df['score'].max() - middle_east_df['score'].min()
cv_middle_east = (range_middle_east/middle_east_mean)*100

print(f"South Asia Coefficient of Variation: {cv_south_asia}")
print(f"Middle East Coefficient of Variation: {cv_middle_east}")
```

```python
# 5. Correlation Analysis:
freedom_corr_south_asia = south_asian_df['Freedom to make life choices'].corr(south_asian_df['score'])
generosity_corr_south_asia = south_asian_df['Generosity'].corr(south_asian_df['score'])

print(f"Correlation with Freedom to Make Life Choices SA: {freedom_corr_south_asia}")
print(f"Correlation with Generosity: {generosity_corr_south_asia}")

freedom_corr_middle_east = middle_east_df['Freedom to make life choices'].corr(middle_east_df['score'])
generosity_corr_middle_east = middle_east_df['Generosity'].corr(middle_east_df['score'])

print(f"Correlation with Freedom to Make Life Choices ME: {freedom_corr_middle_east}")
print(f"Correlation with Generosity: {generosity_corr_middle_east}")

metrics = ['Freedom to make life choices', 'Generosity']
# 2. Create scatter plots with trendlines

for metric in metrics:
    plt.figure(figsize=(8, 5))
    sns.regplot(x=metric, y='score', data=south_asian_df,label='south asian countries')
    sns.regplot(x=metric, y='score', data=middle_east_df,label='middle east countries',color= 'red')
    plt.title(f"Scatter Plot : {metric} vs Score")
    plt.xlabel(metric)
    plt.ylabel('Score')
    plt.grid(axis='y',linestyle='--')
    plt.legend()
    plt.show()

plt.figure(figsize=(12,6 ))
sns.heatmap(south_asian_df[['Log GDP per capita', 'Social support', 'Healthy life expectancy','score','Perceptions of corruption','Dystopia + residual','Generosity','Freedom to make life choices']].corr(), annot=True, )
plt.title('Heatmap of South Asian Region')
plt.show()

plt.figure(figsize=(12,6 ))
sns.heatmap(middle_east_df[['Log GDP per capita', 'Social support', 'Healthy life expectancy','score','Perceptions of corruption','Dystopia + residual','Generosity','Freedom to make life choices']].corr(), annot=True, )
plt.title('Heatmap of Middle East Region')
plt.show()
```
